## 1 Disease Sampling 

To see the how diseases are dicussed and mentioned in Twitter, the first step is to identify what diseases, or say what diseases hashtags to, to explore.

We suppose that user in Twitter would mainly discuess about diseases that cause more deaths in their own countris since those dieases could be more relevant to them， would have higher media exposure and they could have more chance to get in touch with thoes diseases. So to better understand disease dicussion in Twitter, diseases that cause higher deaths in countries having more Twitter users could be the proper sample.
Based on a Twitter users intestigation , 'Countries with the most Twitter users 2019',  published on Sstatista.com, America, Japan, Russia, United Kingdom, Saudi Arabia, Turkey, Brazil, India, Mexico and Spain are the top 10 countries having the highest numbers of Twitter users where as of October 2019. 

And the number of deaths caused by diesase are collect from a dataset of annual number of deaths by cause of each countries in 'Our world in data'. This dataset present the annual death amounts caused by most frequent and common causes of all the countries around the world from 1990 to 2017. 

Data of Twitter's top 10 user's countries were extracted by calculating the average deaths number of each cause in each country from 1900 to 2017. Then all average the deaths number of each cause in each country were added to to get a general distribution of death number of each cause in the top 10 Twitter users' countries. All the causes data were further sorted by death numbers. Aftering removing un-disease causes from the dataset, top 10 diseases were choose as sample whose name serve as Hashtags further for Twitter data collection.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('annual-number-of-deaths-by-cause.csv')

# select the top 10 countries
country = ['United States','Japan','Russia','United Kingdom','Saudi Arabia','Turkey','Brazil','India','Mexico','Spain']

# count the average death numbers caused by each death cause in these 10 counties
dataset = dataset[dataset['Entity'].isin(country)].groupby(by='Entity').mean()
dataset.to_csv('top 10 countries death number data.csv',encoding='utf-8' )
dataset
```

```python
# count the total death number caused by each death cause
dataset.sum().to_csv('average death num is top 10.csv',encoding='utf-8')
death_sum = pd.read_csv('average death num is top 10.csv')
death_sum = death_sum.rename(columns={'Year':'Cause of Death','20035.0':'Death Number'})
```

```python
# sort the dataframe by death number
death_sum = death_sum.sort_values(by='Death Number',ascending=False)
death_sum = death_sum.reset_index(drop=True)
death_sum.to_csv('final list.csv',encoding='utf-8')
death_sum
```
